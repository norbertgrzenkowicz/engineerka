```python
def n_codes(fnames, is_partial=True):
  "Gather the codes from a list of `fnames`"
  vals = set()
  if is_partial:
    random.shuffle(fnames)
    fnames = fnames[:10]
  for fname in fnames:
    msk = np.array(PILMask.create(fname))
    for val in np.unique(msk):
      if val not in vals:
        vals.add(val)
  vals = list(vals)
  p2c = dict()
  for i,val in enumerate(vals):
    p2c[i] = vals[i]
  return p2c
```

```python
p2c = n_codes(lbl_names)
```

```python
p2c
```

```python
def get_msk(fn, p2c):
  fn = path_lbl/f'{fn.stem}{fn.suffix}'
  msk = np.array(PILMask.create(fn))
  mx = np.max(msk)
  for i, val in enumerate(p2c):
    msk[msk==p2c[i]] = val
  return PILMask.create(msk)
```

```python
codes = ['lane', 'not lane']

name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['not lane']
name2id
```

```python
def cam_acc(inp, targ):
  targ = targ.squeeze(1)
  mask = targ != void_code
  return (inp.argmax(dim=1)[mask]==targ[mask]).float().mean()
```

```python
opt = ranger
```

```python
get_y = lambda o: get_msk(o, p2c)
```

```python
binary = DataBlock(blocks=(ImageBlock, MaskBlock(codes)),
                   get_items=get_image_files,
                   splitter=RandomSplitter(),
                   get_y=get_y,
                   item_tfms=Resize((250, 480)),
                   batch_tfms=[Normalize.from_stats(*imagenet_stats)])

                #        batch_tfms=[*aug_transforms(size=(448,448), do_flip=True, flip_vert=True, max_rotate=180, max_zoom=1.3, p_affine = 0.9, p_lighting = 0.9), 
                # Normalize.from_stats(*imagenet_stats)]
```

```python
path_im
```

```python
dls = binary.dataloaders(path_im, bs=2)
```

```python
dls.show_batch(cmap='Purples', vmin=0, vmax=1)
```

```python
learn = unet_learner(dls, resnet34, metrics=cam_acc, self_attention=True, act_cls=Mish, opt_func=opt)
```
