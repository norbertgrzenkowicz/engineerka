```python
%matplotlib ipympl
import matplotlib.cbook as cbook
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL

from mpl_interactions import image_segmenter
```

```python
dataset_path = '/home/norbert/Documents/datasets/roads2/'
dataset_classes = ("curvy", "straight")
data_paths = []
for classes in dataset_classes: data_paths.append(dataset_path + classes)


data_paths[0]
```

```python
segmentedDataSetPath = '/home/norbert/Documents/datasets/segmentedRoads'

def saveSegmentedData(data_class, segmentedSegmenter, counter):
    new_data = PIL.Image.fromarray(segmentedSegmenter.mask * 255)
    if new_data.mode != 'RGB':
        new_data = new_data.convert('RGB')
    
    identifier = 'R' if data_class == 'curvy' else 'r'    

    saved_path = segmentedDataSetPath + '/' + data_class + '/' +  identifier + str(counter) + '.jpg'

    print(saved_path)

    new_data.save(saved_path)

    return saved_path
```

```python
counter = 0

curvy_photos = os.listdir(data_paths[0])
straight_photos = os.listdir(data_paths[1])
```

```python

data_path = data_paths[0] + '/' + curvy_photos[counter]
print(data_path)
img = (PIL.Image.open(data_path)).resize((128, 128))
image = plt.imread(data_path)
segmenter = image_segmenter(image, mask_colors="white", mask_alpha=0.76, figsize=(7, 7))
display(segmenter)
```

```python
print("Saved segmented picture to  {0}".format(saveSegmentedData("curvy", segmenter, counter)))

if len(os.listdir("/home/norbert/Documents/datasets/segmentedRoads/curvy")) < len(curvy_photos):
    counter += 1
```

```python

data_path = data_paths[1] + '/' + straight_photos[counter]
print(data_path)
img = (PIL.Image.open(data_path)).resize((128, 128))
image = plt.imread(data_path)
segmenter = image_segmenter(image, mask_colors="white", mask_alpha=0.76, figsize=(7, 7))
display(segmenter)
```

```python
print("Saved segmented picture to  {0}".format(saveSegmentedData("straight", segmenter, counter)))

if len(os.listdir("/home/norbert/Documents/datasets/segmentedRoads/straight")) < len(straight_photos):
    counter += 1
```
